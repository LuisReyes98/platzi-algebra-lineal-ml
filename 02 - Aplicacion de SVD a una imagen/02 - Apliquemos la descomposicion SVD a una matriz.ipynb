# Apliquemos la descomposición SVD a una imagen

````python
%matplotlib inline

import numpy as np

import matplotlib.pyplot as plt
from PIL import Image
````

```python
imagen_gr = imagen.convert('LA')
```

Convierte la imagen RGB a escala de grises, pero entrega una imagen con 2 canales(o bandas) L y A:

L tiene la información de la imagen en escala de grises.

A tiene la información de transparencia de la imagen, como estamos trabajando con una imagen JPG todos los valores en esta matriz son 255.

```python
imagen_mat = np.array(list(imagen_gr.getdata(band=0)),float)
```

Aquí estamos tomando la información del canal(banda) L, que es el que es la que tiene la información que nos importa, la información de la imagen en escala de grises.


````python
plt.style.use('classic')
imagen = Image.open("./imagenes/imagen_ejemplo_frida_bredesen.jpg")
plt.imshow(imagen)
````

````python
imagen_gr = imagen.convert('LA')

imagen_matrix = np.array(list(imagen_gr.getdata(band=0)), float)

imagen_matrix.shape = (imagen_gr.size[1], imagen_gr.size[0])
````

````python
U, D, V = np.linalg.svd(imagen_matrix)
````

````python
print(imagen_matrix.shape)
print(U.shape)
print(D.shape) # valores singulares
print(V.shape)
````

````python
# los valores singulares son tal que los primeros valores son los que tienen mas informacion y mayor varianza de los datos
````

````python
imagen_recons = np.matrix(U[:,:1] * np.diag(D[:1]) * np.matrix(V[:1,:]))
imagen_recons
````

````python
plt.imshow(imagen_recons, cmap='gray')
````

# Buscando la cantidad de valores singulares que nos sirvan
